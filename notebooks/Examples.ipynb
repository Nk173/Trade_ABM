```python
import networkx as nx
import matplotlib.transforms as mtransforms


def plot_trade_network(trade_matrices, cnames, title=None, tariffs=None, exporter=None, importer=None):
    fig, axes = plt.subplots(1, len(trade_matrices), figsize=(6 * len(trade_matrices), 6))

    if len(trade_matrices) == 1:
        axes = [axes]  # Ensure axes is iterable for a single subplot

    G_base = nx.DiGraph()
    for name in cnames:
        G_base.add_node(name)
    pos = nx.spring_layout(G_base, seed=42, k=10)  # Nodes closer together

    base_matrix = trade_matrices[0]

    for idx, (ax, trade_matrix) in enumerate(zip(axes, trade_matrices)):
        G = G_base.copy()

        # Normalize based on max from base scenario
        max_trade = base_matrix.values.max()

        edge_colors = []
        edge_widths = []
        curved_edges = []

        if idx == 0:
            # Plot full edges in base scenario
            for i, exporter_ in enumerate(cnames):
                for j, importer_ in enumerate(cnames):
                    if trade_matrix.iloc[i, j] > 0:
                        weight = trade_matrix.iloc[i, j] / max_trade
                        G.add_edge(exporter_, importer_)
                        edge_colors.append('black')
                        edge_widths.append(weight * 5)
                        curved_edges.append((exporter_, importer_))

                            # rate = tariffs[importer, exporter, :].mean()
            ntitle= title[0]               
            ax.set_title(ntitle)

        else:
            # Plot differences
            diff = trade_matrix - base_matrix
            for i, exporter_ in enumerate(cnames):
                for j, importer_ in enumerate(cnames):
                    change = diff.iloc[i, j]
                    if change != 0:
                        weight = (abs(change)/diff.sum().sum()) * 0.1
                        color = 'green' if change > 0 else 'red'
                        G.add_edge(exporter_, importer_)
                        edge_colors.append(color)
                        edge_widths.append(weight)
                        curved_edges.append((exporter_, importer_))

            if title is not None:
                # rate = tariffs[importer, exporter, :].mean()

                ax.set_title(title[idx])
            else:
                title = "Difference"
                ax.set_title(title)

        # Draw nodes first
        nx.draw_networkx_nodes(G, pos, node_size=1000, node_color='skyblue', edgecolors='black', linewidths=1, ax=ax, alpha=0.6)
        nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold', ax=ax)

        # Then draw edges (they’ll appear on top)
        nx.draw_networkx_edges(
            G, pos,
            edgelist=curved_edges,
            width=edge_widths,
            edge_color=edge_colors,
            connectionstyle='arc3,rad=0.2',
            arrows=True,
            ax=ax
        )
        
        # Add subplot label
        label = f"({chr(97 + idx)})"  # Generates '(a)', '(b)', '(c)', etc.
        trans = mtransforms.ScaledTranslation(-40/72, -5/72, fig.dpi_scale_trans)
        ax.text(1.0, 1.0, label, transform=ax.transAxes + trans,
                verticalalignment='top', fontsize=20, va='top', ha='left', weight='bold')

    plt.tight_layout()
    plt.show()
# Example usage with trade matrices from results
# grouped_matrix_1 = results[5].groupby(level=0).sum()
# grouped_matrix_2 = results_wtariff[5].groupby(level=0).sum()
# plot_trade_network([grouped_matrix_1, grouped_matrix_2], cnames)

# 
# grouped_matrix_1 = results_cal_0[5].groupby(level=0).sum()
# grouped_matrix_2 = results_cal_tar05[5].groupby(level=0).sum()
# plot_trade_network([grouped_matrix_1, grouped_matrix_2], cnames_cal, 'Tariffs on CHN by US')

# 
grouped_matrix_1 = results_cal_0[5].groupby(level=0).sum()
grouped_matrix_2 = results_cal_tar1[5].groupby(level=0).sum()
plot_trade_network([grouped_matrix_1, grouped_matrix_2], cnames_cal, 'Tariffs on CHN by US')

# grouped_matrix_1 = results_cal_0[5].groupby(level=0).sum()
# grouped_matrix_2 = results_cal_tar2[5].groupby(level=0).sum()
# plot_trade_network([grouped_matrix_1, grouped_matrix_2], cnames_cal, 'Tariffs on CHN by US')

grouped_matrix_1 = results_cal_0[5].groupby(level=0).sum()
grouped_matrix_2 = results_cal_tar_r_1[5].groupby(level=0).sum()
plot_trade_network([grouped_matrix_1, grouped_matrix_2], cnames_cal, 'Reciprocal Tariffs by CHN, US ')

grouped_matrix_0 = results_cal_0[5].groupby(level=0).sum()
grouped_matrix_1 = results_cal_tar1[5].groupby(level=0).sum()
grouped_matrix_2 = results_cal_tar_r_1[5].groupby(level=0).sum()

plot_trade_network(
    [grouped_matrix_0, grouped_matrix_1, grouped_matrix_2],
    cnames_cal, title = ['No Tariffs', 'Tariffs on CHN by US', 'Reciprocal Tariffs by CHN, US']
)

# grouped_matrix_1 = results_cal_0[5].groupby(level=0).sum()
# grouped_matrix_2 = results_cal_tar_r_05[5].groupby(level=0).sum()
# plot_trade_network([grouped_matrix_1, grouped_matrix_2], cnames_cal, 'Reciprocal Tariffs by CHN, US ')
```
